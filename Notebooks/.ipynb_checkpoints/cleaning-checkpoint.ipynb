```python
import pandas as pd
import numpy as np
```

```python
attacks = pd.read_csv("../data/attacks.csv",encoding='cp1252')
```

```python
attacks.head()
```

```python
attacks
```

```python
attacks.info()
```

```python
attacks.Injury
```

```python
attacks['Case Number'].isin(attacks['Case Number.1']).value_counts()
```

```python
# SOlo hay un 10 por ciento de diferencia asi que se puede eliminar la columna y no me interes mucho esta columna
attacks_= attacks.drop(['Case Number'], axis = 1)
```

```python
# realizo esta funcion para quitar todas las filas completa de NaN
attacks_1 = attacks_.dropna(how='all')
attacks_1
```

```python
attacks_1['Unnamed: 22']
```

```python
#Hago lo mismo con las columnas
attacks_2 = attacks_1.dropna(axis=1, how='all')
attacks_2
```

```python

attacks_3 = attacks_1
attacks_3
```

```python
# verifico si son parecidas
attacks_3['Case Number.1'].isin(attacks_3['Case Number.2']).value_counts()
```

```python
# Elimino las columnas que no me interesan
attacks_4 = attacks_3.drop(['Unnamed: 22', 'Unnamed: 23', 'Name', 'pdf', 'href formula', 'href', 'Case Number.1', 'Case Number.2', 'original order'], axis = 1)
attacks_4
```

```python
attacks_4 = attacks_4.dropna(how='all')
attacks_4
```

```python
attacks_4 = attacks_4.fillna(0)
attacks_4
```

```python

attacks_4['Year'] = attacks_4.Year.astype(np.int64)
attacks_4
```

```python
# quito los espacios de los nombres de las columnas
attacks_4.rename(columns={'Species ': 'Species'}, inplace=True)
attacks_4.columns = attacks_4.columns.str.strip()
```

```python
set(attacks_4['Type'])
```

```python
# Llamo a boating, boat, boatomg de la misma forma
attacks_4['Type'] = attacks_4['Type'].str.replace('Boating','Boat')
attacks_4['Type'] = attacks_4['Type'].str.replace('Boatomg','Boat')
```

```python
set(attacks_4['Activity'])
```
