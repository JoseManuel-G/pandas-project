```python
# SOlo hay un 10 por ciento de diferencia asi que se puede eliminar la columna y no me interes mucho esta columna
attacks_= attacks.drop(['Case Number'], axis = 1)
```

```python
# realizo esta funcion para quitar todas las filas completa de NaN
attacks_1 = attacks_.dropna(how='all')
attacks_1
```

```python
#Hago lo mismo con las columnas
attacks_2 = attacks_1.dropna(axis=1, how='all')
attacks_2
```

```python

attacks_3 = attacks_1
attacks_3
```

```python
# verifico si son parecidas
attacks_3['Case Number.1'].isin(attacks_3['Case Number.2']).value_counts()
```

```python
attacks_3['href formula'].isin(attacks_3['href']).value_counts()
```

```python
# Elimino las columnas que no me interesan o que están duplicadas
attacks_4 = attacks_3.drop(['Unnamed: 22', 'Unnamed: 23', 'Name', 'pdf', 'href', 'Case Number.1', 'original order'], axis = 1)
attacks_4
```

```python
# Vuelvo a eliminar todas las filas que tengan todos los elementos NAN. (por si al borrar columnas surgen mas filas)
attacks_5 = attacks_4.dropna(how='all')
attacks_5
```

```python
attacks_6 = attacks_5.fillna(0)
attacks_6
```

```python

attacks_6['Year'] = attacks_6.Year.astype(np.int64)
attacks_6
```

```python
# quito los espacios de los nombres de las columnas
attacks_6.rename(columns={'Species ': 'Species'}, inplace=True)
attacks_6.columns = attacks_6.columns.str.strip()

```

```python
set(attacks_6['Type'])
```

```python
# Llamo a boating, boat, boatomg de la misma forma
attacks_6['Type'] = attacks_6['Type'].str.replace('Boating','Boat')
attacks_6['Type'] = attacks_6['Type'].str.replace('Boatomg','Boat')
```
